분류모델(LogisticRegression)01에서 구현한 모델로 f1-score 및 precision 구해보기

```python
import pandas as pd

import nltk
from nltk.tokenize import TreebankWordTokenizer

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.metrics import f1_score

from sklearn.preprocessing import Binarizer
import matplotlib.pyplot as plt
```

```python
with open('upcycling_info.txt','r') as f:
    data = f.readlines()
print(data)
```

```python
column_name = data[0].strip().split()

print(column_name)
```

```python
data_split = [x.strip().split() for x in data[1:]]

print(data_split)
```

```python
df = pd.DataFrame(data_split, columns = column_name)

df
```

```python
df['contents'] = df['contents'].str.replace('/', ' ')
df
```

```python
df.info()
```

```python
cols = ['color', 'contents', 'category', 'gender', 'price', 'style']
df['features'] = df[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df.drop(['color', 'contents', 'category', 'gender', 'price', 'style'], axis=1, inplace=True)
df
```

```python
df = df[['features', 'brand']]
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.features, df.brand, test_size=0.2, random_state=4)

print('#Train set size: ', len(X_train))
print('#Test set size: ', len(X_test))
```

```python
tb_tokenizer = TreebankWordTokenizer()

tfidf = TfidfVectorizer(tokenizer=tb_tokenizer.tokenize, max_features=2000)

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

clf = LogisticRegression(max_iter=100)
clf.fit(X_train_tfidf, y_train)

pred = clf.predict(X_test_tfidf)
```

```python
# 오차행렬, 정확도, 정밀도, 재현율을 한번에 출력하기 위한 함수 정의

def get_clf_eval(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred, average='micro')
    recall = recall_score(y_test, pred, average='micro')
    print('Confusion Matrix')
    print(confusion)
    print('정확도:{}, 정밀도:{}, 재현율:{}'.format(accuracy, precision, recall))
```

* 오류 수정  
    Target is multiclass but average='binary'. Please choose another average setting, one of [None, 'micro', 'macro', 'weighted'].  
    : f1, recall, roc_auc 같은 경우에는 이진 분류일 경우에 사용이 가능하다고 함  
    다중분류 모델에서는 average 라는 파라미터값을 설정해주면 된다.  
    
* binary : 기본 값, 오직 이진 분류에서 사용
* micro : 총 true positive, false negative, false positive 를 이용해 스코어를 계산함
* macro : 각 레이블의 매트릭스를 계산하고 각 레이블의 가중이 되지 않는 평균(unweighted mean)을 찾는다. 레이블 불균형을 고려하지 않는다.
* weighted : 각 레이블의 매트릭스를 계산하고, 각 레이블에 대한 실제 인스턴스의 수 별로 가중된 평균을 찾는다. precision 과 recall 사이에 있지 않은 f-score 를 초래할 수 있다.
* samples : 각 인스턴스에 대한 매트릭스를 계산하고, 평균을 찾는다. (오직 accuracy_score와 다른 다중 분류에서만 의미가 있다.)

```python
get_clf_eval(y_test, pred)
```

```python
f1 = f1_score(y_test, pred, average='micro')
print('f1 score :', f1)
```
